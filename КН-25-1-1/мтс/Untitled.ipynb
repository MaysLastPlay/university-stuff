```python
import openpyxl
from openpyxl.styles import Font, Alignment, PatternFill, Border, Side

wb = openpyxl.Workbook()
font_family = "Arial"
header_font = Font(name=font_family, size=11, bold=True, color="FFFFFF")
regular_font = Font(name=font_family, size=10)
bold_font = Font(name=font_family, size=10, bold=True)

primary_fill = PatternFill(start_color="1F497D", end_color="1F497D", fill_type="solid")
zebra_fill = PatternFill(start_color="F2F5F9", end_color="F2F5F9", fill_type="solid")

thin_line = Side(border_style="thin", color="D9D9D9")
thick_bottom = Side(border_style="medium", color="1F497D")

thin_border = Border(left=thin_line, right=thin_line, top=thin_line, bottom=thin_line)
header_border = Border(left=thin_line, right=thin_line, top=thin_line, bottom=thick_bottom)


ws1 = wb.active
ws1.title = "Розрахунки комбінаторики"
ws1.views.sheetView[0].showGridLines = True

ws1.merge_cells("A1:K1")
ws1["A1"] = "Лабораторна робота №2: Автоматизація комбінаторних розрахунків"
ws1["A1"].font = Font(name=font_family, size=16, bold=True, color="1F497D")
ws1["A1"].alignment = Alignment(horizontal="left", vertical="center")
ws1.row_dimensions[1].height = 40

categories = [("A3:C3", "Сполучення (Combinations)"), 
              ("E3:G3", "Розміщення (Permutations)"), 
              ("I3:K3", "Перестановки (Factorials)")]

for cells, text in categories:
    ws1.merge_cells(cells)
    start_cell = cells.split(":")[0]
    ws1[start_cell] = text
    ws1[start_cell].font = Font(name=font_family, size=11, bold=True, color="1F497D")
    ws1[start_cell].alignment = Alignment(horizontal="center", vertical="center")

headers = {
    "A": "n", "B": "r", "C": "C(n, r) [Формула]",
    "E": "n", "F": "r", "G": "A(n, r) [Формула]",
    "I": "n", "J": "Елементи з повт.", "K": "P(n) [Формула]"
}

ws1.row_dimensions[4].height = 25
for col, text in headers.items():
    cell = ws1[f"{col}4"]
    cell.value = text
    cell.font = header_font
    cell.fill = primary_fill
    cell.alignment = Alignment(horizontal="center", vertical="center", wrap_text=True)
    cell.border = header_border

data_rows = [
    {"cn": 10, "cr": 3, "pn": 10, "pr": 3, "fn": 5, "f_rep": ""},
    {"cn": 12, "cr": 5, "pn": 12, "pr": 5, "fn": 7, "f_rep": ""},
    {"cn": 15, "cr": 4, "pn": 15, "pr": 4, "fn": 8, "f_rep": ""},
    {"cn": 20, "cr": 6, "pn": 20, "pr": 6, "fn": 10, "f_rep": ""},
    {"cn": 8,  "cr": 2, "pn": 8,  "pr": 2, "fn": 12, "f_rep": ""},
    {"cn": 30, "cr": 5, "pn": 18, "pr": 3, "fn": 6, "f_rep": "2;2;2"},
    {"cn": 14, "cr": 7, "pn": 14, "pr": 7, "fn": 9, "f_rep": "3;3;3"},
]

start_row = 5
for i, row in enumerate(data_rows):
    r_num = start_row + i
    ws1.row_dimensions[r_num].height = 20
    is_zebra = (i % 2 != 0)
    current_fill = zebra_fill if is_zebra else PatternFill(fill_type=None)

    ws1[f"A{r_num}"] = row["cn"]
    ws1[f"B{r_num}"] = row["cr"]
    ws1[f"C{r_num}"] = f"=COMBIN(A{r_num}, B{r_num})"
    
    ws1[f"E{r_num}"] = row["pn"]
    ws1[f"F{r_num}"] = row["pr"]
    ws1[f"G{r_num}"] = f"=PERMUT(E{r_num}, F{r_num})"
    
    ws1[f"I{r_num}"] = row["fn"]
    ws1[f"J{r_num}"] = row["f_rep"]

    if row["f_rep"]:
        ws1[f"K{r_num}"] = f"=FACT(I{r_num}) / (FACT(2)*FACT(2)*FACT(2))"
    else:
        ws1[f"K{r_num}"] = f"=FACT(I{r_num})"

    for col in ["A", "B", "C", "E", "F", "G", "I", "J", "K"]:
        c = ws1[f"{col}{r_num}"]
        c.font = regular_font
        c.border = thin_border
        if current_fill.fill_type:
            c.fill = current_fill
            
        if col in ["C", "G", "K"]:
            c.font = bold_font
            c.number_format = '#,##0'
            c.alignment = Alignment(horizontal="right", vertical="center")
        else:
            c.alignment = Alignment(horizontal="center", vertical="center")

ws1.column_dimensions["C"].width = 18
ws1.column_dimensions["G"].width = 18
ws1.column_dimensions["K"].width = 22
ws1.column_dimensions["D"].width = 4
ws1.column_dimensions["H"].width = 4

ws2 = wb.create_sheet(title="Методичні вказівки")
ws2.views.sheetView[0].showGridLines = True
ws2["A1"] = "МЕТОДИЧНІ ВКАЗІВКИ ДО ЛАБОРАТОРНОЇ РОБОТИ"
ws2["A1"].font = Font(name=font_family, size=14, bold=True, color="1F497D")
ws2.row_dimensions[1].height = 30

guidelines = [
    ("", ""),
    ("1. Опис функцій комбінаторики в Excel:", ""),
    ("Сполучення", "Рахує кількість способів вибрати r елементів з n без урахування порядку. Синтаксис: =COMBIN(n; r)"),
    ("Розміщення", "Рахує кількість способів вибрати r елементів з n з урахуванням порядку. Синтаксис: =PERMUT(n; r)"),
    ("Перестановки", "Рахує кількість унікальних порядків для n елементів (факторіал). Синтаксис: =FACT(n)"),
    ("", ""),
    ("2. Порядок виконання роботи:", ""),
    ("Крок 1", "Ознайомитися з теоретичним матеріалом та варіантом індивідуального завдання."),
    ("Крок 2", "Заповнити стовпці параметрів n та r відповідними числовими значеннями."),
    ("Крок 3", "У стовпцях результатів прописати відповідні математичні формули Excel, посилаючись на комірки даних."),
    ("Крок 4", "Переконатися, що великі значення відображаються у зручному числовому форматі без експонент."),
    ("Крок 5", "Зробити висновки щодо швидкості зростання комбінаторних значень.")
]

for idx, (title, desc) in enumerate(guidelines, start=2):
    ws2.row_dimensions[idx].height = 20
    if title.endswith(":") or title.startswith("1.") or title.startswith("2."):
        ws2[f"A{idx}"] = title
        ws2[f"A{idx}"].font = Font(name=font_family, size=11, bold=True, color="1F497D")
    else:
        ws2[f"A{idx}"] = title
        ws2[f"B{idx}"] = desc
        ws2[f"A{idx}"].font = bold_font
        ws2[f"B{idx}"].font = regular_font

ws2.column_dimensions["A"].width = 20
ws2.column_dimensions["B"].width = 85

ws3 = wb.create_sheet(title="Шаблон Звіту")
ws3.views.sheetView[0].showGridLines = True
ws3["A1"] = "ЗВІТ З ЛАБОРАТОРНОЇ РОБОТИ №2"
ws3["A1"].font = Font(name=font_family, size=14, bold=True, color="1F497D")

report_fields = [
    ("Студент (ПІБ):", "_____________________________________"),
    ("Група:", "__________"),
    ("Варіант:", "__________"),
    ("Дата виконання:", "__________"),
    ("", ""),
    ("Результати обчислень (динамічні лінки на розрахунки):", ""),
    ("Максимальна кількість сполучень у варіанті:", "='Розрахунки комбінаторики'!C5"), 
    ("Максимальна кількість розміщень у варіанті:", "='Розрахунки комбінаторики'!G5"),
    ("Максимальне значення факторіалу:", "='Розрахунки комбінаторики'!K5"),
    ("", ""),
    ("Висновок:", "У ході виконання лабораторної роботи було освоєно інструментарій табличного процесора MS Excel для роботи з комбінаторними конфігураціями. Було підтверджено, що кількість розміщень перевищує кількість сполучень при однакових параметрах вибірки, оскільки враховує фактор порядку розташування елементів.")
]

for idx, (label, val) in enumerate(report_fields, start=3):
    ws3.row_dimensions[idx].height = 22
    ws3[f"A{idx}"] = label
    ws3[f"B{idx}"] = val
    ws3[f"A{idx}"].font = bold_font
    if "Максимальна" in label or "Максимальне" in label:
        ws3[f"B{idx}"].font = Font(name=font_family, color="1F497D", bold=True)
    else:
        ws3[f"B{idx}"].font = regular_font

ws3.column_dimensions["A"].width = 42
ws3.column_dimensions["B"].width = 60

file_name = "Лабораторна_Робота_2_Комбінаторика.xlsx"
wb.save(file_name)
```
