### Короткі методичні вказівки та розрахункові формули

При виконанні роботи в Python замість вбудованих функцій Excel використовуються математичні інструменти стандартного модуля `math`:

1. **Сполучення (Combinations) bez povtoreń:** Кількість способів вибрати $r$ елементів з множини в $n$ елементів без врахування порядку.
   $$C_n^r = \frac{n!}{r!(n-r)!}$$
   * В Python реалізується як: `math.comb(n, r)`

2. **Розміщення (Permutations) bez povtoreń:** Кількість впорядкованих підмножин з $r$ елементів, вибраних з $n$-елементної множини.
   $$A_n^r = \frac{n!}{(n-r)!}$$
   * В Python реалізується як: `math.perm(n, r)`

3. **Перестановки (Factorials) bez povtoreń:** Кількість способів впорядкувати множину з $n$ елементів.
   $$P_n = n!$$
   * В Python реалізується як: `math.factorial(n)`

4. **Перестановки з повтореннями:** Якщо в множині з $n$ елементів деякі з них повторюються $k_1, k_2, ..., k_m$ разів, то кількість унікальних перестановок дорівнює:
   $$P_n(k_1, k_2, ..., k_m) = \frac{n!}{k_1! \cdot k_2! \cdot ... \cdot k_m!}$$
   * В Python обчислюється шляхом ділення загального факторіалу на добуток факторіалів повторюваних підмножин.

### Хід роботи та виконання розрахунків

Нижче наведено скрипт, який використовує реальні вхідні дані вашого варіанту з `lab2.xlsx`, автоматично проводить математичні обчислення для всіх рядків та формує підсумкову аналітичну таблицю (DataFrame).

```python
import math
import pandas as pd

# 1. Вхідні дані з лабораторної роботи (згідно з вашим excel-файлом)
data = [
    {"n_comb": 12,  "r_comb": 5, "n_perm": 12,  "r_perm": 5, "n_fact": 8,   "repeats": []},
    {"n_comb": 9,   "r_comb": 7, "n_perm": 9,   "r_perm": 7, "n_fact": 8,   "repeats": []},
    {"n_comb": 124, "r_comb": 6, "n_perm": 124, "r_perm": 6, "n_fact": 6,   "repeats": []},
    {"n_comb": 56,  "r_comb": 2, "n_perm": 56,  "r_perm": 2, "n_fact": 8,   "repeats": [7, 3]},
    {"n_comb": 4,   "r_comb": 2, "n_perm": 4,   "r_perm": 2, "n_fact": 9,   "repeats": []},
    {"n_comb": 8,   "r_comb": 4, "n_perm": 8,   "r_perm": 4, "n_fact": 30,  "repeats": []},
    {"n_comb": 6,   "r_comb": 2, "n_perm": 6,   "r_perm": 2, "n_fact": 6,   "repeats": []},
    {"n_comb": 7,   "r_comb": 3, "n_perm": 7,   "r_perm": 3, "n_fact": 8,   "repeats": []},
    {"n_comb": 2,   "r_comb": 1, "n_perm": 2,   "r_perm": 1, "n_fact": 20,  "repeats": [7, 4, 9]},
    {"n_comb": 9,   "r_comb": 2, "n_perm": 9,   "r_perm": 2, "n_fact": 5,   "repeats": []}
]

processed_rows = []

# 2. Розрахунковий цикл
for i, row in enumerate(data, start=1):
    # Обчислення комбінацій C(n, r)
    c_res = math.comb(row["n_comb"], row["r_comb"])
    
    # Обчислення розміщень A(n, r)
    a_res = math.perm(row["n_perm"], row["r_perm"])
    
    # Обчислення перестановок P(n) з урахуванням можливих повторень
    if row["repeats"]:
        denom = 1
        for k in row["repeats"]:
            denom *= math.factorial(k)
        p_res = math.factorial(row["n_fact"]) // denom
        rep_str = ";".join(map(str, row["repeats"]))
    else:
        p_res = math.factorial(row["n_fact"])
        rep_str = "-"
        
    # Збереження результатів
    processed_rows.append({
        "Рядок": f"№{i}",
        "n (C)": row["n_comb"],
        "r (C)": row["r_comb"],
        "C(n, r)": c_res,
        "n (A)": row["n_perm"],
        "r (A)": row["r_perm"],
        "A(n, r)": a_res,
        "n (P)": row["n_fact"],
        "Повторення": rep_str,
        "P(n)": p_res
    })

# 3. Візуалізація результатів через Pandas DataFrame
df = pd.DataFrame(processed_rows)
pd.set_option('display.float_format', lambda x: '%.0f' % x)  # Прибираємо наукову нотацію для великих чисел
df.set_index("Рядок", inplace=True)
df
```

Це пояснюється тим, що у розміщеннях критично важливим є порядок розташування об'єктів у підмножині, тоді як у сполученнях оцінюється виключно унікальний компонентний склад.

### Контрольні питання:
1. **Що описує комбінаторика?** 
   *Відповідь:* Розділ математики, що вивчає властивості та способи підрахунку кількості можливих комбінацій, перестановок, розміщень елементів у скінченних множинах.
2. **Чим відрізняється розміщення від сполучення?** 
   *Відповідь:* У розміщеннях порядок елементів у вибірці має значення (наприклад, {1,2} та {2,1} — це два різні розміщення). У сполученнях порядок значення не має, враховується лише склад (наприклад, {1,2} та {2,1} — це одне й те саме сполучення).
3. **Що таке перестановки з повтореннями?** 
   *Відповідь:* Це розміщення елементів множини у певному порядку, коли серед цих елементів є однакові (тотожні) об'єкти. Оскільки перестановка однакових елементів між собою не створює нової комбінації, загальний факторіал ділиться на факторіали кількості повторень кожного елемента.
4. **Яка функція Python використовується для розрахунку факторіалу?** 
   *Відповідь:* Для розрахунку факторіалу використовується вбудована функція `math.factorial(n)` з бібліотеки `math`.
