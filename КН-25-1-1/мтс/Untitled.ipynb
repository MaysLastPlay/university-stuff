### Короткі методичні вказівки та розрахункові формули

*При виконанні роботи в Python замість вбудованих функцій Excel використовуються математичні інструменти стандартного модуля `math`:*

*1. **Сполучення (Combinations) без повторень:** Кількість способів вибрати $r$ елементів з множини в $n$ елементів без врахування порядку.*
   $$C_n^r = \frac{n!}{r!(n-r)!}$$
   * *В Python реалізується як:* `math.comb(n, r)`

*2. **Розміщення (Permutations) без повторень:** Кількість впорядкованих підмножин з $r$ елементів, вибраних з $n$-елементної множини.*
   $$A_n^r = \frac{n!}{(n-r)!}$$
   * *В Python реалізується як:* `math.perm(n, r)`

*3. **Перестановки (Factorials) без повторень:** Кількість способів впорядкувати множину з $n$ елементів.*
   $$P_n = n!$$
   * *В Python реалізується як:* `math.factorial(n)`

4. **Перестановки з повтореннями:** Якщо в множині з $n$ елементів деякі з них повторюються $k_1, k_2, ..., k_m$ разів, то кількість унікальних перестановок дорівнює:
   $$P_n(k_1, k_2, ..., k_m) = \frac{n!}{k_1! \cdot k_2! \cdot ... \cdot k_m!}$$
   * *В Python обчислюється шляхом ділення загального факторіалу на добуток факторіалів повторюваних підмножин.*

### Хід роботи та виконання розрахунків

#### ЗАВДАННЯ 1: Розрахунок кількості сполучень (Combinations) без повторень
##### Пояснення: Сполучення використовуються тоді, коли ми обираємо групу елементів, і їхній внутрішній порядок нам абсолютно НЕВАЖЛИВИЙ (наприклад, вибір делегації).
###### Формула: C = n! / (r! * (n - r)!)
*Беремо параметри з нашого масиву: n = 12, r = 5*

```python
import math

n_comb = 12
r_comb = 5

c_res = math.comb(n_comb, r_comb)

print(f"[Завдання 1] Вхідні дані: n = {n_comb}, r = {r_comb}")
print(f"Пояснення розв'язку: Оскільки порядок вибору не має значення, застосовуємо формулу сполучень.")
print(f"Результат C({n_comb}, {r_comb}) = {c_res} способів.\n")
```

#### Завдання 2: Розрахунок кількості розміщень (Permutations) без повторень
##### Пояснення: Розміщення потрібні тоді, коли ми не просто обираємо елементи, а кожному з них призначаємо певну роль/позицію (ПОРЯДОК елементів МАЄ ЗНАЧЕННЯ).
###### Формула: A = n! / (n - r)!
*Беремо параметри для порівняння: n = 12, r = 5*

```python
import math

n_comb = 12
r_comb = 5

# Замість math.perm(n_comb, r_comb) пишемо універсальну формулу:
a_res = math.factorial(n_comb) // math.factorial(n_comb - r_comb)

print(f"Результат A({n_comb}, {r_comb}) = {a_res}")
```

#### ЗАВДАННЯ 3: Розрахунок перестановок із повтореннями
##### Пояснення: Якщо серед усіх елементів (n) є групи однакових об'єктів, то їхня перестановка між собою не створює нових варіантів. Тому загальний факторіал n! потрібно поділити на добуток факторіалів кількості цих повторень.
###### Формула: P = n! / (k1! * k2!)
*Беремо дані з нашого масиву: n = 8, елементи повторюються 7 та 3 рази.*

```python
n_fact = 8
repeats = [7, 3]

# Крок 1: Знаходимо знаменник — добуток факторіалів усіх повторюваних груп
denom = 1
for k in repeats:
    denom *= math.factorial(k)  # Множимо між собою 7! та 3!

# Крок 2: Ділимо загальний факторіал n! на отриманий знаменник.
# Використовуємо оператор '//' (ділення націло), щоб отримати чисте ціле число (int),
# оскільки кількість комбінацій у математиці не може бути дробовим числом.
p_res = math.factorial(n_fact) // denom

print(f"[Завдання 3] Вхідні дані: n = {n_fact}, групи повторень = {repeats}")
print(f"Пояснення розв'язку: Знаходимо чистий факторіал n={n_fact}! та ділимо на добуток факторіалів повторень ({repeats}).")
print(f"Результат перестановок з повтореннями P = {p_res}.")
```
